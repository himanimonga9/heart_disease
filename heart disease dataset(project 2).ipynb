```python
sns.countplot(x='restecg', data=df, palette=["orange","seagreen"])
plt.xlabel('resting electrocardiographic results')
plt.ylabel('Count')
plt.title('Distribution of resting electrocardiographic')
plt.show()
```

```python
sns.countplot(x='exang', data=df , palette =["darkblue", "grey"])
plt.xlabel('Exercise induced angina')
plt.ylabel('Count')
plt.title('Distribution of exercise induced angina')
plt.show()
```

```python
#0: Represents the absence of exercise-induced angina. This indicates that the individual did not experience chest pain during physical activity.
#1: Represents the presence of exercise-induced angina. This indicates that the individual experienced chest pain  during physical activity.
# The majority of people do not experience chest pain during exercise.
```
